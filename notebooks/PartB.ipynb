Sources of Prediction Error

Class Imbalance (from EDA): ~65% No-Diabetes vs 35% Diabetes → model tends to predict the majority class, resulting in higher false negatives (missed diabetes cases).
Data Quality Issues: Even after imputation and outlier removal, the original dataset contained many zeros (especially Insulin and SkinThickness), which introduced noise.
Limited Features: Only 8 clinical variables; missing important factors such as family history text notes, lifestyle data, or HbA1c levels.
Model Limitations: Logistic Regression assumes linear relationships; non-linear interactions (e.g., Glucose × BMI) are not fully captured.

Practical Implications in Healthcare (Patient Intelligence)

False Negatives (missed high-risk patients) are the most dangerous → a diabetic patient may not receive timely intervention, leading to complications (e.g., retinopathy, kidney failure) and higher long-term costs.
False Positives → unnecessary tests or medication, increasing hospital workload and patient anxiety.
Real-World Deployment: The model should be used as a risk-scoring tool (not final diagnosis) with a lower decision threshold (e.g., probability > 0.4) to increase recall and catch more at-risk patients.
Recommendation: Combine with textual clinical notes (Part C) and retrain periodically to reduce model drift.

Overall, while the model provides valuable predictive intelligence, it must be integrated carefully into clinical workflows with human oversight to minimise patient harm.
